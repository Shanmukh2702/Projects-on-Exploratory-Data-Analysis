```python
# deleting the customer id and resturant id because there is no use to calculate the profitable analysis
df=df.drop(['Customer ID'],axis=1)
df=df.drop(['Restaurant ID'],axis=1)
```

```python
df
```

```python
# Convert 'Order Date and Time' and 'Delivery Date and Time' to datetime
df['Order Date and Time'] = pd.to_datetime(df['Order Date and Time'])
df['Delivery Date and Time'] = pd.to_datetime(df['Delivery Date and Time'])
```

```python
df.isnull().sum()
# Discounts and Offers  has missing values 
```

```python
# fill the missing values 0,
df['Discounts and Offers'].fillna(0,inplace=True)
```

```python
# Calculating net revenue for each order
df['Net Revenue'] = df['Order Value'] - df['Discounts and Offers'] - df['Delivery Fee'] - df['Payment Processing Fee']
```

```python
# Check the dataset after adding this new column
print(df[['Order Value', 'Discounts and Offers', 'Delivery Fee', 'Payment Processing Fee', 'Net Revenue']].head(7))
```

```python
# data visualization
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


plt.figure(figsize=(10,6))
sns.histplot(df['Order Value'], bins=30, kde=True)
plt.title('Distribution of Order Value')
plt.xlabel('Order Value')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Payment Processing Fee'], bins=30, kde=True)
plt.title('Distribution of Order Value')
plt.xlabel('Payment Processing Fee ')
plt.ylabel('Frequency')
plt.show()
```

```python
# Aggregate revenue by day
df['Order Date'] = df['Order Date and Time'].dt.date
daily_revenue = df.groupby('Order Date')['Net Revenue'].sum().reset_index()

plt.figure(figsize=(12,6))
sns.lineplot(data=daily_revenue, x='Order Date', y='Net Revenue', marker="o")
plt.title('Daily Net Revenue Trend')
plt.xticks(rotation=45)
plt.ylabel('Net Revenue')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Payment Method', y='Net Revenue', data=df)
plt.title('Net Revenue by Payment Method')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Discounts and Offers', y='Net Revenue', data=df)
plt.title('Impact of Discounts on Net Revenue')
plt.xlabel('Discounts and Offers')
plt.ylabel('Net Revenue')
plt.show()
```
